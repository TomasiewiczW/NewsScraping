```python
# import

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from datetime import date
import time
from bs4 import BeautifulSoup
import pandas as pd
import requests
import os
import re
```

```python
# hyperlinks

jwShortLinks = ['https://www.jw.org/pl/nauki-biblijne/rodzina/randki-chodzenie-ze-soba/', \
               'https://www.jw.org/pl/nauki-biblijne/rodzina/gospodarowanie-pieniedzmi/', \
               'https://www.jw.org/pl/nauki-biblijne/rodzina/wymiana-mysli/',]

jwLongLinks = ['https://www.jw.org/pl/nauki-biblijne/rodzina/bozy-poglad-malzenstwo/', \
               'https://www.jw.org/pl/nauki-biblijne/rodzina/wychowywanie-dzieci/', \
               'https://www.jw.org/pl/nauki-biblijne/rodzina/wychowywanie-nastolatkow/']
links = []
```

```python
page = requests.get(jwShortLinks[0])
soup = BeautifulSoup(page.content)
article_container = soup.find('article')
print(article_container['h1'])
```

```python
for j in jwShortLinks:
    page = requests.get(j)
    soup = BeautifulSoup(page.content)
    article_container = soup.find('div', class_ = 'sortByResultsGrouping')
    article_body = article_container.find_all('a')
    i = 0
    for ab in article_body:
        i += 1
        if i%2 == 0:
            l = "https://www.jw.org" + ab['href']
            links.append(l)
            print("https://www.jw.org" + ab['href'])
    print("Number of articles: " + str(len(links)))
```

```python
for j in jwLongLinks:
    page = requests.get(j)
    soup = BeautifulSoup(page.content)
    article_container = soup.find_all('div', class_ = 'pageSectionContainer')
    i = 0
    for ac in article_container:
        article_body = ac.find_all('a')
        for ab in article_body:
            i += 1
            if i%2 == 0:
                l = "https://www.jw.org" + ab['href']
                links.append(l)
                print("https://www.jw.org" + ab['href'])
    print("Number of articles: " + str(len(links)))

```

```python
t = 0

for l in links:
    page = requests.get(l)
    soup = BeautifulSoup(page.content)
    print('_____________________________________________________________')
    article_container = soup.find('div', class_ = 'bodyTxt')
    article_body = article_container.find_all('p')
    for i in article_body:
        print(i.text)
        with open('0_4_'+ str(t) + ".txt", "w", encoding="utf-8") as f:
            f.write(i.text)            
    t += 1






```

```python
a = 0

for l in links:
    page = requests.get(l)
    soup = BeautifulSoup(page.content)
    article_container = soup.find('article').find('h1')

    articleFull = article_container.text + " "
    articleNoBrackets = articleFull
    articleBracketsOnly = ""

    article_container = soup.find('div', class_ = 'bodyTxt')
    article_body = article_container.find_all('p')
    for i in article_body:
        articleFull += (i.text + " ")
        brackets = re.findall('\(.*?\)',articleFull)
        if len(brackets) > 0:
            for b in brackets:
                articleBracketsOnly += (b + " ")
        articleNoBrackets = re.sub("\(.*?\)","",articleFull)

    with open('articleFull' + str(links.index(l)) + ".txt", "w", encoding="utf-8") as f:
        f.write(articleFull)    
    with open('articleNoBrackets' + str(links.index(l)) + ".txt", "w", encoding="utf-8") as f:
        f.write(articleNoBrackets)    
    with open('articleBracketsOnly' + str(links.index(l)) + ".txt", "w", encoding="utf-8") as f:
        f.write(articleBracketsOnly)    
    
    print('\r' + str(links.index(l)) + '/' + str(len(links)) + ' completed', end='', flush=True)

        
    
    
```
